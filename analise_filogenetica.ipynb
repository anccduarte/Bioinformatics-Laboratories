## Alinhamentos múltiplos e filogenia

```python
from Bio import Phylo
```

```python
files = [["rrnB","dna"],["ampC","dna"],["ampC","prot"]]
```

```python
for file in files:
    print(f"----> Árvore filogenética relativa a {file[0]} ({file[1]}) <----")
    print("")
    tree = Phylo.read(f"clustalw_{file[0]}_{file[1]}.dnd","newick")
    Phylo.draw_ascii(tree)
    print("\n")
```
