## AnÃ¡lise de literatura

```python
from Bio import Entrez
from Bio import Medline
```

```python
term1 = "Escherichia coli rrnB tetracycline resistance"
term2 = "Escherichia coli ampC beta-lactamase"
terms = [term1,term2]
```

```python
for term in terms:
    print(f"----> Artigos relacionados a '{term}' <----\n")
    # --> procurar o nÃºmero de artigos que contÃªm os termos indicados <--
    Entrez.email = "pg45464@alunos.uminho.pt"
    handle = Entrez.egquery(term=term)
    record = Entrez.read(handle)
    for row in record["eGQueryResult"]:
        if row["DbName"] == "pubmed":
            num_artigos = row["Count"]
    handle.close()
    # --> listar os IDs dos artigos que contÃªm os termos indicados <--
    Entrez.email = "pg45464@alunos.uminho.pt"
    handle = Entrez.esearch(db="pubmed",term=term,retmax=num_artigos)
    record = Entrez.read(handle)
    idlist = record["IdList"]
    handle.close()
    # --> listar o tÃ­tulo, autores e source dos artigos encontrados anteriormente <--
    Entrez.email = "pg45464@alunos.uminho.pt"
    handle = Entrez.efetch(db="pubmed",id=idlist,rettype="medline",retmode="text")
    record = Medline.parse(handle)
    records = list(record)
    for record in records:
        print("Title:", record.get("TI", "?"))
        print("Authors:", ", ".join(record.get("AU", "?")))
        print("Source:", record.get("SO", "?"))
        print("")
    print("")
    handle.close()
```
